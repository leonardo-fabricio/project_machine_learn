```python
import pandas as pd
df = pd.read_csv("random_sales_products.csv")

df
```

```python
df.plot(kind="scatter", x="date", y ="sales");
```

```python
# Separando dias meses e anos
df["date"] = pd.to_datetime(df["date"], format="%d/%m/%Y")
df["day"] = df["date"].dt.day
df["month"] = df["date"].dt.month
df["year"] = df["date"].dt.year

df
```

```python
df["product_code"] = df["product"].astype("category").cat.codes #criando um codigo para cada produto
```

```python
# create model 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


x = df[["day", "month", "year", "product_code"]]
y = df["sales"]

x_train, x_test, y_train, y_test  = train_test_split(x, y, test_size=0.2, random_state=42)


model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Avaliação
from sklearn.metrics import mean_squared_error

y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
print(f"MSE do modelo: {mse}")
```
